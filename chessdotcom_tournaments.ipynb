```python
import json
import pandas as pd

from urllib import request
from bs4    import BeautifulSoup
```

```python
url = "https://api.chess.com/pub/tournament/vii-arena-loggichess-1171620/1"
data = request.urlopen(url).read()
soup = BeautifulSoup(data,'html.parser')
site_json=json.loads(soup.text)
site_json
```

```python
# url
url = [d.get('url') for d in site_json['games'] if d.get('url')]

# time_control
time_control = [d.get('time_control') for d in site_json['games'] if d.get('time_control')]

# end_time
end_time = [d.get('end_time') for d in site_json['games'] if d.get('end_time')]
 
# rated
rated = [d.get('rated') for d in site_json['games'] if d.get('rated')]

# fen
fen = [d.get('fen') for d in site_json['games'] if d.get('fen')]

# time_class
time_class = [d.get('time_class') for d in site_json['games'] if d.get('time_class')]

# rules
rules = [d.get('rules') for d in site_json['games'] if d.get('rules')]

# eco
eco = [d.get('eco') for d in site_json['games'] if d.get('eco')]

#create games dataframe
df_games = pd.DataFrame(data=[url,time_control,end_time,rated,fen,time_class,rules,eco]).T
df_games.columns = ['url',
                    'time_control',
                    'end_time',
                    'rated',
                    'fen',
                    'time_class',
                    'rules',
                    'eco']

df_games.head()
```

```python
# white players
white = [d.get('white') for d in site_json['games'] if d.get('white')]
df_white_players = pd.DataFrame(data=white)
df_white_players.columns = ['white_rating','white_result','white_id','white_username']

# black players
black = [d.get('black') for d in site_json['games'] if d.get('black')]
df_black_players = pd.DataFrame(data=black)
df_black_players.columns = ['black_rating','black_result','black_id','black_username']

# merge white and black players dataframes
df_players = df_white_players.merge(df_black_players, left_index=True, right_index=True)
df_players.head()
```

```python
# create pgn dataframe
pgn = [d.get('pgn') for d in site_json['games'] if d.get('pgn')]
df_pgn = pd.DataFrame(data=pgn)
df_pgn.columns = ['pgn']

# split columns
df_pgn = df_pgn['pgn'].str.split("\n",expand=True)
df_pgn.columns = ['event',
                  'site',
                  'date',
                  'round',
                  'white',
                  'black',
                  'result',
                  'current_position',
                  'timezone',
                  'eco',
                  'eco_url',
                  'utc_date',
                  'utc_time',
                  'white_elo',
                  'black_elo',
                  'time_control',
                  'termination',
                  'start_time',
                  'end_date',
                  'end_time',
                  'link',
                  'col_1',
                  'moves',
                  'col_2']

# drop empty cols
cols_drop = ['col_1','col_2']
df_pgn = df_pgn.drop(cols_drop, axis=1)

df_pgn.head().T
```

```python
# pgn data cleaning

# dataframe cleaning
df_pgn = df_pgn.apply(lambda x: x.replace('"','',regex=True))
df_pgn = df_pgn.apply(lambda x: x.replace('\]','',regex=True))
df_pgn = df_pgn.apply(lambda x: x.replace('\[','',regex=True))

# event
df_pgn['event'] = df_pgn['event'].apply(lambda x: x.replace('Event ',''))

# site
df_pgn['site'] = df_pgn['site'].apply(lambda x: x.replace('Site ',''))

# date
df_pgn['date'] = df_pgn['date'].apply(lambda x: x.replace('Date ',''))
df_pgn['date'] = pd.to_datetime(df_pgn['date'], format = '%Y-%m-%d').dt.date.astype('datetime64')

# round
df_pgn['round'] = df_pgn['round'].apply(lambda x: x.replace('Round ',''))

# white
df_pgn['white'] = df_pgn['white'].apply(lambda x: x.replace('White ',''))

# black
df_pgn['black'] = df_pgn['black'].apply(lambda x: x.replace('Black ',''))

# result
df_pgn['result'] = df_pgn['result'].apply(lambda x: x.replace('Result ',''))

# current_position
df_pgn['current_position'] = df_pgn['current_position'].apply(lambda x: x.replace('CurrentPosition ',''))

# timezone
df_pgn['timezone'] = df_pgn['timezone'].apply(lambda x: x.replace('Timezone ',''))

# eco
df_pgn['eco'] = df_pgn['eco'].apply(lambda x: x.replace('ECO ',''))

# eco_url
df_pgn['eco_url'] = df_pgn['eco_url'].apply(lambda x: x.replace('ECOUrl ',''))

# utc_date
df_pgn['utc_date'] = df_pgn['utc_date'].apply(lambda x: x.replace('UTCDate ',''))
df_pgn['utc_date'] = pd.to_datetime(df_pgn['utc_date'], format = '%Y-%m-%d').dt.date.astype('datetime64')

# utc_time
df_pgn['utc_time'] = df_pgn['utc_time'].apply(lambda x: x.replace('UTCTime ',''))
df_pgn['utc_time'] = pd.to_datetime(df_pgn['utc_time'], format = '%H:%M:%S').dt.time

# white_elo
df_pgn['white_elo'] = df_pgn['white_elo'].apply(lambda x: x.replace('WhiteElo ',''))
df_pgn['white_elo'] = df_pgn['white_elo'].astype(float)

# black_elo
df_pgn['black_elo'] = df_pgn['black_elo'].apply(lambda x: x.replace('BlackElo ',''))
df_pgn['black_elo'] = df_pgn['black_elo'].astype(float)

# time_control
df_pgn['time_control'] = df_pgn['time_control'].apply(lambda x: x.replace('TimeControl ',''))

# termination
df_pgn['termination'] = df_pgn['termination'].apply(lambda x: x.replace('Termination ',''))

# start_time
df_pgn['start_time'] = df_pgn['start_time'].apply(lambda x: x.replace('StartTime ',''))
df_pgn['start_time'] = pd.to_datetime(df_pgn['start_time'], format = '%H:%M:%S').dt.time

# end_date
df_pgn['end_date'] = df_pgn['end_date'].apply(lambda x: x.replace('EndDate ',''))
df_pgn['end_date'] = pd.to_datetime(df_pgn['end_date'], format = '%Y-%m-%d').dt.date.astype('datetime64')

# end_time
df_pgn['end_time'] = df_pgn['end_time'].apply(lambda x: x.replace('EndTime ',''))
df_pgn['end_time'] = pd.to_datetime(df_pgn['end_time'], format = '%H:%M:%S').dt.time

# link
df_pgn['link'] = df_pgn['link'].apply(lambda x: x.replace('Link ',''))

# moves
df_pgn['moves'] = df_pgn['moves'].apply(lambda x: x.replace('%clk',''))
```
